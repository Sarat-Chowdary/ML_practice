```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
```

```python
df=pd.read_excel('Data_SVM.xlsx', dtype=float)
```

```python
Y_origin=df['Y']
feat=df.drop('Y',axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8), facecolor='yellow')
plt.scatter(x=df['X1'][0], 
                        y=df['X2'][0], 
                        c='g',
                        s=20, label='Class 1' 
                        )
plt.scatter(x=df['X1'][0], 
                        y=df['X2'][0], 
                        c='b',
                        s=20, label='class -1' 
                        )

for i in range(len(df)):
    if df['Y'][i]==-1:
        plt.scatter(x=df['X1'][i], 
                        y=df['X2'][i], 
                        c='g',
                        s=20
                        )
    else:
        plt.scatter(x=df['X1'][i], 
                        y=df['X2'][i], 
                        c='b',
                        s=20 
                        )
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
clf = svm.SVC(kernel='linear', C=2)
clf.fit(feat, Y_origin)
```

```python
x_test = np.array([[8,15],[7,4]])
# array.reshape(1,-1)
clf.predict(x_test)
```

```python
w = clf.coef_[0]       
b = clf.intercept_[0]      
x_points = np.linspace(0, 16)    
y_points = -(w[0] / w[1]) * x_points - b / w[1]  
w_hat = clf.coef_[0] / (np.sqrt(np.sum(clf.coef_[0] ** 2)))
margin = 1 / np.sqrt(np.sum(clf.coef_[0] ** 2))
svm_line = np.array(list(zip(x_points, y_points)))
margin_top = svm_line + w_hat * margin
margin_bottom = svm_line - w_hat * margin
```

```python
from sympy import Point, Line
  
p1, p2, p3 = Point(0, 0), Point(2, 3), Point(-2, 2)
  
l1 = Line(p1, p2)
l2 = l1.perpendicular_line(p3)

p11 = Point(0,y_points[0])
p12 = Point(16,y_points[-1])
p1 = Point(clf.support_vectors_[0, 0],
            clf.support_vectors_[0, 1])

p2 = Point(clf.support_vectors_[2, 0],
            clf.support_vectors_[2, 1])

p3 = Point(clf.support_vectors_[5, 0],
            clf.support_vectors_[5, 1])

p4 = Point(clf.support_vectors_[6, 0],
            clf.support_vectors_[6, 1])

l = Line(p11,p12)

l1 = l.perpendicular_line(p1)
l2 = l.perpendicular_line(p2)
l3 = l.perpendicular_line(p3)
l4 = l.perpendicular_line(p4)

p1 = l.intersection(l1)
p2 = l.intersection(l2)
p3 = l.intersection(l3)
p4 = l.intersection(l4)

print(p1,p2,p3,p4)
```
