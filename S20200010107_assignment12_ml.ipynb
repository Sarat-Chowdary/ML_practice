```python
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
train, test = mnist.load_data()
```

```python
x_train = train[0]
y_train = train[1]
x_test = test[0]
y_test = test[1]
```

```python
i=4
plt.imshow(x_train[i])
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.models import Sequential
```

```python
model = Sequential([
        Flatten(input_shape=(28,28)),
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(10, activation='softmax'),
    ])

model.compile(optimizer = 'adam',
              loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, validation_split=0.2, epochs=20, verbose=1)
model.evaluate(x_test, y_test)
```

### This model consists of an initial flatten layer and 2 hidden layers - first dense layer consisting of 64 nodes the second one consits of 32 nodes and both use the activation function - 'RELU'.
<br>


### The last layer - output layer consists of 10 nodes for 10 class classification. learning rate is set to default (Tensorflow default) while huperparameter tuning.

```python
accuracy = [0.001,0.003,0.004,0.008,0.01]
```

```python
def plotting(history):
    plt.figure(facecolor='grey')
    l1 = plt.plot(history.history['loss'])
    l2 = plt.plot(history.history['val_loss'])
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    labels = ["Loss", "Validation Loss"]
    plt.legend([l1, l2], labels=labels,
           loc="upper right")
```

```python
def keras_model():
        model = Sequential([
        Flatten(input_shape=(28,28)),
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(10, activation='softmax'),
    ])
        return model
```

```python
def fitting(lr):
    model = keras_model()
    model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=lr),
                  loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, validation_split=0.2, epochs=20, verbose=1)
    model.evaluate(x_test, y_test)
    return history
```

```python
history = []
for i in range(len(accuracy)):
    history.append(fitting(accuracy[i]))
```

```python
for i in range(len(history)):
    plotting(history[i])
```

## Learning rate of 0.001 seems to be the best followed by 0.003 and 0.004 when compared with respect to validation losses.
<br>

## Learning rate of 0.008 and 0.01 seem to give very bad results without converging. (model diverges).
